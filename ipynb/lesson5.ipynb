## Операторы работы с циклом: continue/break

Схема применения и пример:

```
while условие_окончаия_1:
    if условие_пропуска_итерации:
        continue
    # блок кода
    # ...
    if условие_окончания_2:
        break
    # еще код
    # ...
```

````python
N = 50
n0 = 2
print(f'Все простые числа от {n0} до {N}:')

# Проверяем все числа от n0 до N включительно
for n in range(n0, N+1):
    # флаг проверки делимости
    is_composite = False
    # проверяем делимость на все числа до n
    for i in range(n0, n):
        is_composite = (n % i == 0)
        # достаточно найти хотя бы один делитель
        if is_composite:
            break
    # если число составное, пропускаем его
    if is_composite:
        continue
    print(n, end='  ')
````
