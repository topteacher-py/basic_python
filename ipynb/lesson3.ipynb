# Операторы ветвлений. Часть 1

if/elif/else - связка операторов ветвлений

if - обязательный, всегда первый

elif - необязательный, всегда после if

else - необязательный, всегда последний

if и elif проверяют выполнение логических условий:

```
if условие1:
    # блок кода
    # ...
    # ...
elif условие2:
    # блок кода
    # ...
    # ...
elif условие3:
    # ...
else:
    # финальный блок кода
    # ...
    # ...
```

Блок кода - это некоторое количество строк кода, сдвинутых на один Tab,

при этом после конца текущего блока отступ так же исчезает.

После каждого оператора ветвления идет новый блок кода, но каждый следующий

оператор ветвления начинается с той же строки, что и первый.

````python
# Пример проверки делимости числа
number = 37

# Первое условие: number % 2 == 0
if number % 2 == 0:
    print(f'Число {number} четное')
elif number % 3 == 0:  # Проверка второго условия
    print(f'Число {number} кратно 3, но не кратно 2')
elif number % 5 == 0:  # Проверка третьего условия
    print(f'Число {number} кратно 5, но не кратно 2 и 3')
else:  # Ни одно из условий не прошло
    print(f'Число {number} не кратно ни 2, ни 3, ни 5')
````
