## Пример работы с циклом for

Выведем на экран "ёлочку" с помощью символов "*"

```python
N = 5  # количество итераций (прогонов)
for i in range(N):  # range - диапазон для цикла
    n = 2 * i + 1
    s = (2 * N - n) // 2
    print(f"{' ' * s}{'*' * n}")
```

## Пример работы с циклом while

Выведем на экран цифры заданного числа в обратном порядке

```python
number = 12379
while number > 0:  # проверка условия окончания цикла
    print(number % 10)
    number = number // 10
```
