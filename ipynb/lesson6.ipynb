# Исключения

## Примеры работы с исключениями

Общий пример кода:

```
try:
    # блок кода
except исключение:
    # обработка исключения
```

````python
# обработка любых исключений
try:
    a = 1 / 0
except Exception as ex:
    print(ex)
````

````python
# обработка конкретного исключения
try:
    a = 1 / 0
except ZeroDivisionError:
    print('Поймано исключение при делении на ноль')
````

````python
# пропуск обработки исключения
try:
    a = 1 / 0
except ZeroDivisionError:
    pass
````

## Пример обработки исключения при вводе

````python
n = input('Введите любое целое число: ')

try:
    n = int(n)
    print(f'Вы ввели: {n}')
except ValueError:
    print(f'Ошибка: {n} - не является целым числом')
````

## Пример обработки нескольких исключений

````python
x = 3
y = 1
try:
    print(f'Попробуем разделить {x} на {y}')
    result = x / y
except ZeroDivisionError:
    print("Ошибка: деление на 0")
except TypeError:
    print("Ошибка: неверный тип данных")
else:
    print(f"Результат деления: {result}")
finally:
    print("Строка, которая выводится в любом случае")
````
