## Переменная

```python
# date - переменная
# = - оператор присваивания
# '26 марта' -  значение переменной
date = '26 марта'
print('Сегодня', date)
date = '27 марта'
print('Завтра будет', date)
```

## Сохранение результата ввода в переменную

```python
# input() - приглашение на ввод с клавиатуры
name = input('Введите имя: ')
print('Здравствуйте,', name)
```

## Ввод чисел

```python
day = int(input('Какое сегодня число?\n'))
print('Сегодня', day, 'марта')
print('Завтра будет', day+1, 'марта')
```
