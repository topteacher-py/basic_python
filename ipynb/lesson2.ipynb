# Операторы

## Арифметические






```
+  - сложение
-  - вычитание
*  - умножение
/  - деление
// - деление нацело
%  - остаток от деления
** - возведение в степень
```

Решите примеры с помощью арифметических операторов и выведите результаты:
* 2 + 2 = ?
* 2 х 2 = ?
* 2024^2 = ?
* Какие бывают отстатки от деления на 7?
* Сколько полных недель в году?
* Савелий положил в банк 100 000 рублей под 10% годовых, проценты начисляются на счет каждый месяц, какая сумма будет на счете Савелия через год?

````python
result1 = 2 + 2
print('2 + 2 =', result1)
````

## Операторы присваивания



```
=  - присваивание
+= - присваивание со сложением
-= - присваивание с вычитанием
*= - присваивание с умножением
...
```
Переделайте предыдущее задание с новыми операторами

````python
result1 = 2
result1 += 2
print('2 + 2 =', result1)
````

!= - не равно (не равно ли?)
